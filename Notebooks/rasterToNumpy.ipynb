## Converting raster file to numpy file

Input for this file is a collection of raster tiles which is converted into numpy files with longitude, latitude and elevation.

```python
#Import Libraries
import glob
import gdal
import numpy as np
```

```python
#Provide folder location of the TIFF files
filepattern = 'Split/' 
```

```python
#Collecting Tiff files from folder
impath = "%s*.tif"%(filepattern)
flist=glob.glob(impath)
```

```python
#Converting raster data into 3 dimensional numpy file with Longitude, latitude and elevation
for index, fpath in enumerate(flist):
    ds = gdal.Open(fpath, GA_ReadOnly )
    geotransform = ds.GetGeoTransform()
    raster = ds.ReadAsArray()
    raster[raster < 0] = 0
    ulc = [geotransform[0] , geotransform[3]]    
    fname = 'x-%.6f_y-%.6f.npy'%(ulc[0], ulc[1])
    coordinate_X, coordinate_Y = ulc[0], ulc[1]
    x = np.linspace(coordinate_X, coordinate_X + raster.shape[0], raster.shape[0])
    y = np.linspace(coordinate_Y, coordinate_Y - raster.shape[1], raster.shape[1])
    xx, yy= np.meshgrid(x, y)
    xx = xx.reshape(-1)
    yy = yy.reshape(-1)
    z_data = raster
    z = z_data.reshape(-1)
    coordinates = np.zeros((raster.shape[0]*raster.shape[1], 3))
    coordinates[:,0] = xx
    coordinates[:,1] = yy
    coordinates[:,2] = z
    print fname
    np.save(fname, coordinates)
```
